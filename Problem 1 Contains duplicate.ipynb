##### Problem 1 - Contains Duplicate


```python
# solution 1 - Brute Force Solution.
# Time Complexity - O(n^2)
# Space Complexity. - O(1)
from typing import List
class Solution:
  def hasDuplicate(self, nums : List[int]) -> bool:
    size = len(nums)
    for i in range(size):
      for j in range(i + 1, size):
        if nums[i] == nums[j]:
          return True
    return False

if __name__ == '__main__':
  solution = Solution()

  # Test Case 1.
  nums = [1, 2, 3, 3]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))

  # Test Case 2.
  nums = [1, 2, 3, 4]
  # Output: false - The integer array has no duplicate.
  print(solution.hasDuplicate(nums))

  # Test Cases 3.
  nums = [9, 8, 5, 3, 3, 1, 5, 7]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))
```

```python
# Solution 2 - Sorting Solution.
# Time Complexity - O(nlogn) Sorting + O(n) Passing the Array = O(n log n)
# Space complexity - O(1) .sort(), O(n) sorted
from typing import List
class Solution:
  # sorted() and .sort() -> tim sort -> O(nlog n)
  def hasDuplicate(self, nums : List[int]) -> bool:
    nums.sort()
    size = len(nums)
    for i in range(1, size):
      if nums[i] == nums[i - 1]:
        return True
    return False
  def hasDuplicateSorted(self, nums : List[int]) -> bool:
    nums = sorted(nums)
    size = len(nums)
    for i in range(size - 1):
      if nums[i] == nums[i + 1]:
        return True
    return False

if __name__ == '__main__':
  solution = Solution()

  # Test Case 1.
  nums = [1, 2, 3, 3]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))

  # Test Case 2.
  nums = [1, 2, 3, 4]
  # Output: false - The integer array has no duplicate.
  print(solution.hasDuplicate(nums))

  # Test Cases 3.
  nums = [9, 8, 5, 3, 3, 1, 5, 7]
  # [1, 3, 3, 5, 5, 7, 8, 9]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))
```

```python
# Solution 3 - Hash set - O(1) lookup.
# Time Complexity - O(n) Pass by the array only onec, O(1) lookup in HashSet.
# Space Complexity - O(n) Storing the numbers in the HashSet.
from typing import List
class Solution:
  def hasDuplicate(self, nums : List[int]) -> bool:
    seen = set()
    for num in nums:
      if num in seen:
        return True
      seen.add(num)
    return False

if __name__ == '__main__':
  solution = Solution()

  # Test Case 1.
  nums = [1, 2, 3, 3]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))

  # Test Case 2.
  nums = [1, 2, 3, 4]
  # Output: false - The integer array has no duplicate.
  print(solution.hasDuplicate(nums))

  # Test Cases 3.
  nums = [9, 8, 5, 3, 3, 1, 5, 7]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))
```

```python
# Solution 4 - Hash set length. Sets does not allow duplicates.
# Time Complexity - O(n) - Creating the set requires one pass through the list.
# Space Complexity - O(n) - set(nums) → creates a new set with up to n elements → O(n) space.
from typing import List
class Solution:
  def hasDuplicate(self, nums : List[int]) -> bool:
    return len(set(nums)) < len(nums)

if __name__ == '__main__':
  solution = Solution()

  # Test Case 1.
  nums = [1, 2, 3, 3]
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))

  # Test Case 2.
  nums = [1, 2, 3, 4]
  # Output: false - The integer array has no duplicate.
  print(solution.hasDuplicate(nums))

  # Test Cases 3.
  nums = [9, 8, 5, 3, 3, 1, 5, 7]
  # set(nums) -> {9,8,5,3,1,7}
  # Output: true - The integer array has a duplicate.
  print(solution.hasDuplicate(nums))
```
